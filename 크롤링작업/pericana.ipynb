```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import json
```

```python
import time
seed = np.random.randint(100)
np.random.seed(seed)
a = np.random.randint(5)
time.sleep(1)

pericana = []
cnt = 1

for i in range (1, 111):
    url = f'https://pelicana.co.kr/store/stroe_search.html?page={i}&branch_name=&gu=&si='
    res = requests.get(url)
    res.encoding = 'utf-8'
    html = BeautifulSoup(res.text, 'html.parser')
    
    print(f'{cnt}번째 크롤링 중입니다')
    cnt += 1
    
    
    for j in range(len(html.select('td:nth-child(2)'))):
        pericana.append(html.select('td:nth-child(2)')[j].text)
```

```python
pericana = set(pericana)
len(pericana) # 서대문구 지점 하나만 중복으로 들어가 있어서 set으로 중복 제거 후 다시 list로 변경 했습니다.
```

```python
pericana = list(pericana)
len(pericana)
```

```python
new_pericana = []

for i in range(len(pericana)):
    pericana[i].split(' ')[:2]
    new_pericana.append(pericana[i].split(' ')[:2])
```

```python
df = pd.DataFrame(new_pericana, columns=['addr'])

df['category'] = np.nan
df['brand'] = np.nan
df['category'] = df['category'].fillna('치킨')
df['brand'] = df['brand'].fillna('페리카나')

df = df[['category', 'brand', 'addr']]
df
```

```python
df.to_csv('pericana.csv', index = False)
```
