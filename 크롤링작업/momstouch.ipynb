```python
momstouch = []
cnt = 1

num_list = ['01', '02', '03', '04', '05', '06', '07','08', '09', '10', '11', '12', '13', '14', '15', '16', '17']


for i in num_list:

    url = f'https://momstouchdev.co.kr/store/inner_shop_list.php?s_area_sido=0{i}&s_area_sigun=&type=area'

    info = {

        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36',
        'Referer': 'https://momstouchdev.co.kr/store/main.php'
    }

    response = requests.get(url, headers = info)
    html = BeautifulSoup(response.text, 'html.parser')
    
    print(f'{cnt}번째 크롤링 중입니다')
    cnt += 1
    
    
    for j in range(len(html.select('dd'))):
        momstouch.append(html.select('dd')[j].text)
        
```

```python
new_momstouch = []

for i in range(len(momstouch[::3])):
    sp = []
    sp = momstouch[::3][i].split()
    
    if sp[0] == '경기':
        sp[0] = '경기도'
        
    elif sp[0] == '경남':
        sp[0] = '경상남도'
        
    elif sp[0] == '경북':
        sp[0] = '경상북도'
        
    elif sp[0] == '충남':
        sp[0] = '충청남도'
        
    elif sp[0] == '충북':
        sp[0] = '충청북도'
        
    elif sp[0] == '전남':
        sp[0] = '전라남도'
        
    elif sp[0] == '전북':
        sp[0] = '전라북도'
        
    elif sp[0] == '인천':
        sp[0] = '인천광역시'
    
    elif sp[0] == '광주':
        sp[0] = '광주광역시'
    
    elif sp[0] == '대구':
        sp[0] = '대구광역시'
        
    elif sp[0] == '대전':
        sp[0] = '대전광역시'
        
    elif sp[0] == '부산':
        sp[0] = '부산광역시'
        
    elif sp[0] == '서울':
        sp[0] = '서울특별시'
        
    elif sp[0] == '울산':
        sp[0] = '울산광역시'
        
    elif sp[0] == '제주도':
        sp[0] = '제주특별자치도'
        
    elif sp[0] == '제주':
        sp[0] = '제주특별자치도'
        
    elif sp[0] == '세종':
        sp[0] = '세종특별자치시'    
        
    elif sp[0] == '세종시':
        sp[0] = '세종특별자치시'
        
        
        
        
        
        
##########################################################################        
        
        
    elif sp[1] == '한라대학로':
        sp[1] = '제주시'
    
    
    
    
        
    
    new_momstouch.append(sp[0] + " " + sp[1])
```

```python
np.unique(new_momstouch)
```

```python
df = pd.DataFrame(new_momstouch, columns=['addr'])

df['category'] = np.nan
df['brand'] = np.nan
df['category'] = df['category'].fillna('햄버거')
df['brand'] = df['brand'].fillna('맘스터치')

df = df[['category', 'brand', 'addr']]
df
```

```python
df.to_csv('./momstouch.csv', encoding='utf-8-sig', index = False)
```
