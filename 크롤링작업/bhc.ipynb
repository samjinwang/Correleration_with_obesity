```python
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
import json
```

```python
import time
seed = np.random.randint(100)
np.random.seed(seed)
a = np.random.randint(5)
time.sleep(1)

bhc = []
cnt = 1

for i in range (1, 163):
    url = f'https://www.bhc.co.kr/location/map_sub.asp?cPage={i}&area=&gugubn=&search_text='
    info = {
        'Referer': 'https://www.bhc.co.kr/location/map.asp',
        'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/94.0.4606.61 Safari/537.36'
    }
    res = requests.post(url, headers = info)
    html = BeautifulSoup(res.text, 'html.parser')
    
    print(f'{cnt}번째 크롤링 중입니다')
    cnt += 1
    
    
    for j in range(len(html.select('p.mt10'))):
        bhc.append(html.select('p.mt10')[j].text.strip())
```

```python
bhc[0].split('\n')[0].split(':')[1]
```

```python
addr = []

for i in range(len(bhc)):
    addr.append(bhc[i].split('\n')[0].split(':')[1])
    
addr[0]
```

```python
new_bhc = []

for i in range(len(addr)):
    sp = []
    sp = addr[i].split()
    
    if sp[0] == '경기':
        sp[0] = '경기도'
        
    elif sp[0] == '경남':
        sp[0] = '경상남도'
        
    elif sp[0] == '경북':
        sp[0] = '경상북도'
        
    elif sp[0] == '충남':
        sp[0] = '충청남도'
        
    elif sp[0] == '충북':
        sp[0] = '충청북도'
        
    elif sp[0] == '전남':
        sp[0] = '전라남도'
        
    elif sp[0] == '전북':
        sp[0] = '전라북도'
        
    elif sp[0] == '인천':
        sp[0] = '인천광역시'
    
    elif sp[0] == '광주':
        sp[0] = '광주광역시'
    
    elif sp[0] == '대구':
        sp[0] = '대구광역시'
        
    elif sp[0] == '대전':
        sp[0] = '대전광역시'
        
    elif sp[0] == '부산':
        sp[0] = '부산광역시'
        
    elif sp[0] == '서울':
        sp[0] = '서울특별시'
        
    elif sp[0] == '울산':
        sp[0] = '울산광역시'
        
    elif sp[0] == '제주도':
        sp[0] = '제주특별자치도'
        
    elif sp[0] == '제주':
        sp[0] = '제주특별자치도'
        
    elif sp[0] == '세종':
        sp[0] = '세종특별자치시'    
        
    elif sp[0] == '세종시':
        sp[0] = '세종특별자치시'
        
        
        
        
        
        
##########################################################################        
        
        
    elif sp[1] == '한라대학로':
        sp[1] = '제주시'
    
    
    
    
        
    
    new_bhc.append(sp[0] + " " + sp[1])
```

```python
new_bhc[0]
```

```python
df = pd.DataFrame(new_bhc, columns=['addr'])

df['category'] = np.nan
df['brand'] = np.nan
df['category'] = df['category'].fillna('치킨')
df['brand'] = df['brand'].fillna('BHC')

df = df[['category', 'brand', 'addr']]
df
```

```python
df.to_csv('bhc.csv', index = False)
```
